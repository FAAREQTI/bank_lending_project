```python
# Reading the CSV file into a Pandas DataFrame

file_name = 'loans_full_schema.csv'
data_path = '/content/drive/MyDrive/' + file_name
data = pd.read_csv(data_path)
#a glimpse into the data ...
data.head()
```

```python
# Cleaning the data
import pandas as pd

# Dropping columns with more than 50% NA's
columns_to_remove = data.columns[data.isna().mean() > 0.5]  # Get columns with > 50% missing values
data.drop(columns=columns_to_remove, inplace=True)  # Drop the selected columns
# Replace NA's with "Unknown" in the 'emp_title' column
data['emp_title'].fillna('Unknown', inplace=True)
```

```python
#counting null values in each column
data.isnull().sum()
```

```python
# Clean NA's : for some columns I replaced with mean value as it made a bit of sense (and save time), for another I just deleted lol
# Replace NA values in specified columns with column means
columns_to_fillna = ['months_since_last_credit_inquiry', 'emp_length', 'num_accounts_120d_past_due']

for column in columns_to_fillna:
    data[column].fillna(data[column].mean(), inplace=True)

data.dropna(inplace=True)
data.isnull().sum()
```

```python
# Checking for outliers in numerical columns
def detect_outliers(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = column[(column < lower_bound) | (column > upper_bound)]
    return outliers

# Create a summary table for outliers
outlier_summary = pd.DataFrame(columns=['Column Name', 'Number of Outliers', 'Outlier Values', 'Highest Outlier Value'])

# Select only numerical columns for outlier detection
numerical_columns = data.select_dtypes(include=[np.number])

for column_name in numerical_columns.columns:
    outliers = detect_outliers(numerical_columns[column_name])
    num_outliers = len(outliers)
    outlier_values = outliers.tolist()
    highest_outlier = outliers.max() if not outliers.empty else None
    outlier_summary = outlier_summary.append({'Column Name': column_name,
                                              'Number of Outliers': num_outliers,
                                              'Outlier Values': outlier_values,
                                              'Highest Outlier Value': highest_outlier},
                                             ignore_index=True)

# Display the summary table
print(outlier_summary)

##All outliers seem rather logical and do not point out to a problem in the data, so we will keep them.
```

```python
# Adding an identity key to not loose connection when seperating into tables

data['identity_key'] = data.reset_index().index

# Display the DataFrame with the new identity key
print(data)
```

```python
# Creating relevant tables

# Define the columns for each table
personal_and_financial_columns = [
    'identity_key', 'emp_title', 'emp_length', 'state', 'homeownership',
    'annual_income', 'verified_income', 'debt_to_income', 'balance',
    'paid_total', 'paid_principal', 'paid_interest', 'paid_late_fees'
]

loan_info_columns = [
    'identity_key', 'loan_purpose', 'application_type', 'loan_amount', 'term',
    'interest_rate', 'installment', 'grade', 'sub_grade', 'issue_month',
    'loan_status', 'initial_listing_status', 'disbursement_method'
]

credit_and_delinquency_columns = [
    'identity_key', 'delinq_2y', 'earliest_credit_line', 'inquiries_last_12m',
    'total_credit_lines', 'open_credit_lines', 'total_credit_limit',
    'total_credit_utilized', 'num_collections_last_12m',
    'num_historical_failed_to_pay', 'current_accounts_delinq',
    'total_collection_amount_ever', 'current_installment_accounts'
]

account_and_credit_card_columns = [
    'identity_key', 'accounts_opened_24m', 'months_since_last_credit_inquiry',
    'num_satisfactory_accounts', 'num_accounts_120d_past_due',
    'num_accounts_30d_past_due', 'num_active_debit_accounts',
    'total_debit_limit', 'num_total_cc_accounts', 'num_open_cc_accounts',
    'num_cc_carrying_balance', 'num_mort_accounts', 'account_never_delinq_percent',
    'tax_liens', 'public_record_bankrupt'
]
# Create the "personal_and_financial_transactions_info" table
personal_and_financial_info = data[personal_and_financial_columns].copy()

# Create the "loan_info" table
loan_info = data[loan_info_columns].copy()

# Create the "credit_and_delinquency_info" table
credit_and_delinquency_info = data[credit_and_delinquency_columns].copy()

# Create the "account_and_credit_card_info" table
account_and_credit_card_info = data[account_and_credit_card_columns].copy()

# Display the created tables
print(personal_and_financial_info)
print(loan_info)
print(credit_and_delinquency_info)
print(account_and_credit_card_info)
```
